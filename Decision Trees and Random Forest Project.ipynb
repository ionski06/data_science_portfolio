# Setting up the Data

Let's set up our data for our Random Forest Classification Model.

**Check loans.info() again.**

```python
loans.info()
```

## Categorical Features

Since the **purpose** column as categorical

That means we need to transform them using dummy variables so sklearn will be able to understand them. Will do this in one clean step using pd.get_dummies.

```python
cat_feats = ['purpose']
```

**Now use pd.get_dummies(loans,columns=cat_feats,drop_first=True) to create a fixed larger dataframe that has new feature columns with dummy variables. Set this dataframe as final_data.**

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

## Train Test Split

Now its time to split our data into a training set and a testing set.

** Will use sklearn to split data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Training a Decision Tree Model

** Import DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

**Creating an instance of DecisionTreeClassifier() called dtree and fit it to the training data.**

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predictions and Evaluation of Decision Tree
**Creating predictions from the test set and a classification report and a confusion matrix.**

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## Training the Random Forest model

To test if Random Forest will perform better.

**Creating an instance of the RandomForestClassifier class and fit it to our training data from the previous step.**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

## Predictions and Evaluation

Predicting off the y_test values and evaluate our model.

** Predict the class of not.fully.paid for the X_test data.**

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

**Show the Confusion Matrix for the predictions.**

```python
print(confusion_matrix(y_test,predictions))
```
